```python
# write a function that can return the track record of 2 teams against each other

def winCount(team1, team2):
    relevant_matches = ipl[((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) | ((ipl['Team1'] == team2) & (ipl['Team2'] == team1))]
    count1 = relevant_matches[relevant_matches['WinningTeam'] == team1].shape[0]
    count2 = relevant_matches[relevant_matches['WinningTeam'] == team2].shape[0]

    return count1, count2 
    
```

```python
def winCount(team1, team2):
    relevant_matches = ipl[((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) | ((ipl['Team1'] == team2) & (ipl['Team2'] == team1))]
    count1 = len(relevant_matches[relevant_matches['WinningTeam'] == team1])
    count2 = len(relevant_matches[relevant_matches['WinningTeam'] == team2])

    return count1, count2 
    
```

```python
team1 = 'Chennai Super Kings'
team2 = 'Kolkata Knight Riders'
win1 , win2 = winCount(team1, team2)
print(f'{team1} won {win1} times and {team2} won {win2} times when played against each other.')
```

```python
points_df = pd.DataFrame(
    {
        '1st point': [(3,4), (-6,5), (0,0), (-10, 1), (4, 5)],
        '2nd point':[(-3, 4), (0,0), (2,2), (10,10), (1,1)]
    }
)
```

```python
points_df
```

```python
def euclidean(row):
    pt_A = row['1st point']
    pt_B = row['2nd point']

    return ((pt_A[0] - pt_B[1])**2 + (pt_A[1] - pt_B[1]) **2)** 0.5
```

```python
points_df['distance'] = points_df.apply(euclidean, axis=1)
points_df 
```

```python
# Why separate objects to handle data and time when python already has datetime functionality?
```

- syntax wise datetime is very convenient 
- But the performance takes a hit while working with huge data. List vs Numpy Array 
- The weaknesses of Python's datetime format inspired the NumPy team to add a set of native time series data type to NumPy.
- The datetime64 dtype encodes dates as 64-bit integers, and thus allows arrays of dates to be represented very compactly. 

```python
import numpy as np 
date = np.array('2015-07-04', dtype=np.datetime64)
date 
```

```python
date + np.arange(12)
```

- Beacause of the uniform type in NumPy datetime64 arrays, this type of operation can be accomplished much more quickly than if we were working directly with Python's datetime objects, especially as arrays get large
- Pandas Timestamp object combines the ease-of-use of python datetime with the efficient storage and vectorized interface of numpy.datetime64
- From a group of these Timestamp objects, Pandas can construct a DatetimeIndex that can be used to index data in a Series or DataFrame

```python
data = pd.DataFrame({
    'A': [1, 2, None, 4],
    'B': [10, 20, 30, None],
    'C': [5, None, 15, 20]
})
```

```python
import pandas as pd 
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import StandardScaler

def preprocess_dataset(df):
    # Handle missing value (mean)
    imputer = SimpleImputer(strategy='mean')
    df_imputed = pd.DataFrame(imputer.fit_transform(df), columns = df.columns)

    # Scale the dataset 
    scaler = StandardScaler()
    df_scaled = pd.DataFrame(scaler.fit_transform(df_imputed), columns=df.columns)

    return df_scaled  
```

```python
preprocess_dataset(data)
```

```python
import torch 
import torch.nn as nn 
import torch.optim as optim 

class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(10, 32)
        self.fc2 = nn.Linear(32, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x 
    

model = SimpleNN()
criterion = nn.BCELoss()
optimizer = optim.adam(model.parameters(), lr=0.001)
```
